```python
# prompt: Add /content/Data (1).csv to the map

import plotly.graph_objects as go
import pandas as pd

!pip install plotly
!pip install pandas

# Load the data from the CSV file
df = pd.read_csv('/content/Data (1).csv')

# Assuming your CSV has columns named 'country', 'continent', and 'value'
# Adjust the column names if they are different in your file

# Create the Choropleth map
fig = go.Figure(data=go.Choropleth(
    locations=df['country'],  # Use country names as locations
    z=df['loss_percentage'],  # Color based on 'value' column
    locationmode='country names',  # Set locationmode to 'country names'
    colorscale='Oranges',
    autocolorscale=False,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title='Value',  # Update colorbar title if necessary
))

fig.update_layout(
    title_text='World Map with Continents',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    )
)

fig.show()
```

```python
# prompt: code to plot africa alone not showing unplotted areas given /content/african_data.csv plotting to show loss percentage

import plotly.graph_objects as go
import pandas as pd

# Install necessary libraries if not already installed
!pip install plotly
!pip install pandas

# Load the data from the CSV file
# Ensure the file path is correct
df = pd.read_csv('/content/african_data.csv')

# Filter data for Africa
africa_df = df[df['continent'] == 'Africa']

# Create the Choropleth map
fig = go.Figure(data=go.Choropleth(
    locations=africa_df['country'],
    z=africa_df['loss_percentage'],
    locationmode='country names',
    colorscale='Oranges',
    autocolorscale=False,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title='Loss Percentage',
))


fig.update_layout(
    title_text='Loss Percentage in Africa',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular',
        scope='africa'  # Limit the map to Africa
    ),
)

fig.show()
```
