```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdd = pd.read_csv('/content/drive/MyDrive/numhack/Data.csv')
```

```python
# bar plot for loss percentange v country
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.barplot(x='country', y='loss_percentage', data=pdd) # Replace 'Country' and 'Loss Percentage' with the correct column names if different
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.title('Loss Percentage by Country')
plt.xlabel('country')
plt.ylabel('loss_percentage')
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
east_africa=['Uganda','Burundi','Democratic Republic of the Congo','United Republic of Tanzania','Rwanda','South Sudan','Kenya']

east_african_pdd = pdd[pdd['country'].isin(east_africa)]
east_african_pdd.head()

east_african_pdd.to_csv('east_africa.csv', index = False)
```

```python
plt.figure(figsize=(12, 5))
plt.bar(east_african_pdd['country'], east_african_pdd['loss_percentage'])
plt.xticks(rotation=0, ha='center')  # Rotate x-axis labels for better readability
plt.title('Loss Percentage by Country')
plt.xlabel('country')
plt.ylabel('loss_percentage')
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Calculate the frequency of each food supply stage
stage_counts = east_african_pdd['food_supply_stage'].value_counts()

# Sort the stages by frequency in descending order
sorted_stages = stage_counts.sort_values(ascending=False)

# Print or use the sorted stages
sorted_stages
```

```python
east_africa_stage=['Farm','Harvest','Storage','Transport']

east_african_stage_pdd = east_african_pdd[east_african_pdd['food_supply_stage'].isin(east_africa_stage)]
east_african_stage_pdd.head()
```

```python
# Calculate the frequency of each food supply stage
stage_counts = east_african_pdd['commodity'].value_counts()

# Sort the stages by frequency in descending order
sorted_stages = stage_counts.sort_values(ascending=False)

# Print or use the sorted stages
sorted_stages
```

```python
east_africa_commo=['Maize (corn)','Rice','Millet','Wheat','Sorghum','Barley']

east_african_commo_pdd = east_african_pdd[east_african_pdd['commodity'].isin(east_africa_commo)]
east_african_commo_pdd.head()
```

```python
# Assuming 'east_african_pdd' DataFrame is already created as in the previous code

# Count the frequency of each food supply stage for each country
stage_frequency = east_african_stage_pdd.groupby(['country', 'food_supply_stage']).size().unstack(fill_value=0)

# Create the bar plot
stage_frequency.plot(kind='bar', figsize=(12, 6))
plt.title('Food Supply Stage Frequency per Country (East Africa)')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=0, ha='center')
plt.legend(title='Food Supply Stage')
plt.tight_layout()
plt.show()
```

```python

# Count the frequency of each food supply stage for each country
stage_frequency = east_african_commo_pdd.groupby(['country', 'commodity']).size().unstack(fill_value=0)

# Create the bar plot
stage_frequency.plot(kind='bar', figsize=(12, 6))
plt.title('Food wasted Frequency per Country (East Africa)')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=0, ha='center')
plt.legend(title='Commodity wasted')
plt.tight_layout()
plt.show()
```

### MAIZE PRODUCTION and WASTAGE

```python
maizepdd = pd.read_csv('/content/maize-production.csv')
maizepdd.head()
```

```python
east_africa_maize=['Uganda','Burundi','Democratic Republic of Congo','Tanzania','Rwanda','South Sudan','Kenya']
east_african_maize_pdd = maizepdd[maizepdd['Country'].isin(east_africa_maize)]
east_african_maize_pdd.head()
```

```python
east_african_maize_pdd.info()
```

```python

east_african_maize_pdd.to_csv('east_africa_maize-pdd.csv', index = False)
```

```python
# Group data by country and year, then sum production
production_by_year_country = east_african_maize_pdd.groupby(['Country', 'Year'])['Production (t)'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(10, 5))
for country in production_by_year_country['Country'].unique():
    country_data = production_by_year_country[production_by_year_country['Country'] == country]
    plt.plot(country_data['Year'], country_data['Production (t)'], label=country)

plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Maize Production Over Time by Country (East Africa)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group data by country and year, then sum production
wastage_by_year_country = east_african_maize_pdd.groupby(['Country', 'Year'])['Supply chain waste (t)'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(10, 5))
for country in production_by_year_country['Country'].unique():
    country_data = wastage_by_year_country[wastage_by_year_country['Country'] == country]
    plt.plot(country_data['Year'], country_data['Supply chain waste (t)'], label=country)

plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Maize wastage Over Time by Country (East Africa)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
