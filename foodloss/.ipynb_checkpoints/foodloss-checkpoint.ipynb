```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data2_cleaned.csv')
data.head(3)
```

```python
import pycountry_convert as pc
def get_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        return continent_code
    except KeyError:
        return None

data['continent'] = data['country'].apply(get_continent)
data.head(3)
```

```python
africa_data = data[data['continent']=='AF'].reset_index(drop=True)
africa_data
```

```python
ea_data = africa_data.query("country == 'Burundi' or country == 'Democratic Republic of the Congo' or country == 'Kenya' or country == 'Rwanda' or country == 'Somalia' or country == 'South Sudan' or country == 'Uganda' or country == 'United Republic of Tanzania'").reset_index(drop=True)
ea_data
```

```python
#what are east africa's top five foods?
ea_top_five = ea_data.groupby('commodity')['loss_percentage'].count().sort_values(ascending=False).reset_index()
ea_top_five.head()
```

```python
ea_data_top = ea_data.query("commodity == 'Maize (corn)' or commodity == 'Rice' or commodity == 'Sorghum' or commodity == 'Millet' or commodity == 'Wheat'").reset_index(drop=True)
ea_data_top
```

```python
ea_data_grouped = ea_data.groupby('country')['loss_percentage'].mean().sort_values(ascending=False).reset_index(name='average_loss_percentage')
ea_data_grouped
```

```python
ea_data_top_grouped = ea_data_top.groupby('country')['loss_percentage'].mean().sort_values(ascending=False).reset_index(name='average_loss_percentage')
ea_data_top_grouped
```

```python
import geopandas as gpd
world = gpd.read_file("countries/ne_110m_admin_0_countries.shp")
```

```python
ea_data1 = ea_data_top_grouped.merge(world, left_on='country', right_on='SOVEREIGNT', how='left')
ea_data1
```

```python
geo_ea_data = gpd.GeoDataFrame(ea_data1, geometry='geometry')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
geo_ea_data.plot(
    column='average_loss_percentage', 
    cmap='YlGnBu', 
    legend=True, 
    legend_kwds={
        'label': "Mean Loss Percentage",
        'orientation': "vertical",
        'shrink': 0.5,            
        'pad': 0.02,              
    },
    ax=ax,
    edgecolor='black'
)

ax.set_title('Average Wastage of Top Five Foods from East Africa (2000-2022)', fontsize=12, fontweight='bold')
ax.axis('off')
plt.savefig('eamap1.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
geo_ea_data.plot(
    column='average_loss_percentage', 
    cmap='Blues', 
    legend=True, 
    legend_kwds={
        'label': "Mean Loss Percentage",
        'orientation': "vertical",
        'shrink': 0.5,            
        'pad': 0.02,              
    },
    ax=ax,
    edgecolor='black'
)

ax.set_title('Average Wastage of Top Five Foods from East Africa (2000-2022)', fontsize=12, fontweight='bold')
ax.axis('off')
plt.savefig('eamap2.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
#average wastage of the top five foods by east african countries through years
top_foods_wastage = ea_data_top.groupby(['year', 'commodity'])['loss_percentage'].mean().reset_index()
top_foods_wastage
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=top_foods_wastage, 
             x='year', 
             y='loss_percentage', 
             hue='commodity',
             palette='Blues',
             linewidth=2.0)
plt.title('Average Wastage of Top Five Foods in East Africa from 2000 to 2022 ', fontsize=12, fontweight='bold')
plt.xlabel('Year', fontsize=10, fontweight='bold')
plt.ylabel('Average Loss Percentage (%)', fontsize=10, fontweight='bold')
plt.legend(title='Commodity', fontsize=10)
plt.tight_layout()
plt.savefig('lineplot1.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
#activites that cause most wastage of the maize in east african countries
ea_data_maize = ea_data_top[ea_data_top['commodity'] == 'Maize (corn)'].reset_index()
maize_wastage_activity = ea_data_maize.groupby('activity')['loss_percentage'].count().sort_values(ascending=False).reset_index(name='count')
maize_wastage_activity.head(7)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=maize_wastage_activity.head(3), 
            x='count', 
            y='activity', 
            hue='activity',
            palette='Blues',
            dodge=False,
            legend=False)

plt.title('Top 3 Activities Causing Wastage of Maize in East Africa', fontsize=12, fontweight='bold')
plt.xlabel('Wastage Occurrences', fontsize=10, fontweight='bold')
plt.ylabel('Activity', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig('barplot1.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=maize_wastage_activity.head(3), 
            x='count', 
            y='activity', 
            hue='activity',
            palette='Blues',
            dodge=False,
            legend=False)

plt.title('Top 3 Activities Causing Wastage of Maize in East Africa', fontsize=12, fontweight='bold')
plt.xlabel('Wastage Occurrences', fontsize=10, fontweight='bold')
plt.ylabel('Activity', fontsize=10, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('barplot2.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
#activites that cause most wastage of the wheat in east african countries
ea_data_wheat = ea_data_top[ea_data_top['commodity'] == 'Wheat'].reset_index()
wheat_wastage_activity = ea_data_wheat.groupby('activity')['loss_percentage'].count().sort_values(ascending=False).reset_index(name='count')
wheat_waste = wheat_wastage_activity
wheat_wastage = wheat_waste[wheat_waste['activity'] != 'Drying, Harvesting'].reset_index(drop=True)
wheat_wastage.head(3)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=wheat_wastage.head(3), 
            x='count', 
            y='activity', 
            hue='activity',
            palette='Blues',
            dodge=False,
            legend=False)

plt.title('Top 3 Activities Causing Wastage of Wheat in East Africa', fontsize=12, fontweight='bold')
plt.xlabel('Wastage Occurrences', fontsize=10, fontweight='bold')
plt.ylabel('Activity', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig('barplot3.png', dpi=600, bbox_inches='tight')
plt.show()
```

### Maize Production

```python
maizepdd = pd.read_csv('maize-production.csv')
maizepdd.head()
```

```python
recent_years = maizepdd[maizepdd['Year'] > 1999]
recent_years.head(3)
```

```python
countries = recent_years['Country'].unique()
print(countries)
```

```python
east_africa_maize=['Uganda','Burundi','Democratic Republic of Congo','Tanzania','Rwanda','South Sudan','Kenya']
east_african_maize_pdd = recent_years[recent_years['Country'].isin(east_africa_maize)].reset_index(drop=True)
east_african_maize_pdd
```

```python
production_by_year = east_african_maize_pdd.groupby('Year')['Production (t)'].sum().reset_index()
production_by_year.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=production_by_year, x='Year', y='Production (t)', c='blue', linewidth=2.0)
plt.title('Maize Production by East African Countries from 2000 to 2022 ', fontsize=12, fontweight='bold')
plt.xlabel('Year', fontsize=10, fontweight='bold')
plt.ylabel('Maize Quantity (Kgs)', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig('lineplot_maize.png', dpi=600, bbox_inches='tight')
plt.show()
```
